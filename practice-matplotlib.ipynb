### Matplotlib basic workflow

```python
# Creating two lists (x, y)
x = [30, 44, 60, 50, 81]
y = [9, 22, 37, 49, 91]

# Setup the figure and axes
fig, ax = plt.subplots()

# Add the data to the axes
ax.plot(x, y)

# Customize the plot
ax.set(title="Sample plot",
      xlabel="x-axis",
      ylabel="y-axis");

# Saving the plot
#fig.savefig("simple-plot.png")
```

### Types of plots

### Simple line plot

```python
# Creating an array of 100 numbers, between 0 and 100
x = np.linspace(0, 10, 100)
```

```python
# Creating a plot of x and x ** 2
fig, ax = plt.subplots()
ax.plot(x, x **2);
```

### Scatter plot

```python
# Creating a scatter plot of x and the exponential of x
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Creating a scatter plot of x and sin(x)
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar plot

```python
# Creating a Python dictionary about car sales (USD) per month
car_sales = {"Audi": 2500, "Toyota": 6000, "Mercedes": 7500}
```

```python
# Creating the bar graph - x-axis=keys, y-axis=values
fig, ax = plt.subplots()
ax.bar(car_sales.keys(), car_sales.values())

# Adding a title and labels
ax.set(title="Car Sales per month",
      xlabel="Car Brands",
      ylabel="Total Sales");
```

```python
# Creating a horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(car_sales.keys()),
        list(car_sales.values()))

# Adding a title and labels
ax.set(title="Car Sales per month",
      xlabel="Total Sales",
      ylabel="Car Brands");
```

### Histograms

```python
# Creating a random NumPy array of 1000 numbers using NumPy
x = np.random.randn(1000)

# Creating a histogram plot of X
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplots

```python
# Creating a plot with 2 rows and 2 columns and figsize of (10, 5)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot X versus X/2 on the top left axes
ax1.plot(x, x/2)

# Top right subplot - Scatter plot of 10 random numbers
ax2.scatter(np.random.random(10), np.random.random(10))

# Bottom left subplot - Bar graph of the car sales
ax3.bar(car_sales.keys(), car_sales.values())

# Bottom right subplot - Histogram of 1000 random numbers on the bottom
ax4.hist(np.random.randn(1000));
```

### Plotting data from a DataFrame

```python
# Importing the data
heart_disease = pd.read_csv("dataset-heart-disease.csv")
```

```python
# Reviewing the first 10 rows
heart_disease.head(10)
```

```python
# Histogram fo the 'age' column
heart_disease["age"].plot.hist(bins=50); # bins = number of histogram bins used
```

```python
# Histogram on the DataFrame
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

Visualizing the data on how age and cholesterol combined effect the target in patients over 50 years old.

The 'target' field refers to the presence of heart disease in the patient.
* 0 = no disease.
* 1 = disease.

```python
# Creating a DataFrame with patients over 60 years old
patients_over_60 = heart_disease[heart_disease["age"] > 60]

# Creating the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
scatter = ax.scatter(patients_over_60["age"],
                    patients_over_60["chol"],
                    c=patients_over_60["target"]) # marker colors

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline
ax.axhline(patients_over_60["chol"].mean(),
          linestyle="--");
```

```python
# Using 'cmap' parameters

# Creating the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
scatter = ax.scatter(patients_over_60["age"],
                     patients_over_60["chol"],
                     c=patients_over_60["target"],
                     cmap="winter") # parameter

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Adding a meanline
ax.axhline(patients_over_60["chol"].mean(),
          linestyle="--",
          color="red");
```
