## Solution 2

```python
df
```

```python
# Getting the mean from first column
df_m = df.amount_credit.mean()
df_m
```

```python
# Modify the NaN values from "amount_credit" with the mean we recently get
df["amount_credit"].fillna(value = df_m, inplace = True)
df
```

```python
# Same process with "gender" column but getting the median
df["gender"].fillna(value = df.gender.median(), inplace = True)
df
```

```python
# Getting the mode from "amount_credit" column
df_mo = df.amount_credit.mode()
df_mo
```

```python
# Filling the missing values with the mode
df["amount_credit"].fillna(value = df_mo[0], inplace = True)
df
```

```python
# Checking the missing values
df.isnull().values.any()
```

```python
df
```

```python
# Creating a copy of the original dataframe
df_copy = df.copy()
df_copy
```

```python
# Getting rid of the rows with N/A values
df_copy.dropna(subset=["amount_credit", "gender"], inplace = True)
df_copy
```

```python
# Getting the mode from "amount_credit" column
df_copy.amount_credit.mode()
```

```python
# Generating a dictionary in which we will use the mode and mean for the first columns
dict_1 = {"amount_credit": df_copy.amount_credit.mode()[0], "gender": df_copy.amount_credit.mean()}
print(dict_1)
```

```python
# Filling N/A values
df_copy.amount_credit.fillna(df_copy.amount_credit.mode()[0], inplace = True)
df_copy.gender.fillna(df_copy.mean(), inplace = True)
df_copy
```

What data was considered most important? Why?

The mean as it helped us to fill the missing data.

Was null data removed or replaced? What was done and why?

Data that does not represent a high percentage of the total was deleted. The missing data was filled with the mean.

Is it necessary to sort the data for analysis?

Yes, it is better to first explore the data. Sorting and cleanning the data make the next processes easier.

Are there formatting issues that need to be addressed before the modeling process?
